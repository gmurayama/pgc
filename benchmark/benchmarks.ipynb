```python
from IPython.display import display
import pandas as pd

er_model_addProduct = pd.read_csv('./../data/csv/er_model_addProduct.csv', sep = ';')
er_model_addItemToCart = pd.read_csv('./../data/csv/er_model_addItemToCart.csv', sep = ';')
er_model_showTotal = pd.read_csv('./../data/csv/er_model_showTotal.csv', sep = ';')

display(er_model_addProduct)
display(er_model_addItemToCart)
display(er_model_showTotal)
```

```python
# Add Product

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(25, 12))
plt.plot(er_model_addProduct['gas'], '-o', linewidth=1, markersize=2)
plt.grid()
plt.show()
```

```python
# Add Item to Cart

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(50, 12), dpi=500)
plt.plot(er_model_addItemToCart['gas'], '-o', linewidth=0.1, markersize=0.3)
plt.grid()
plt.show()
```

```python
# Show Total

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20, 8))
plt.plot(er_model_showTotal['gas'], '-o', linewidth=0.5, markersize=1.5)
plt.grid()
plt.show()
```
